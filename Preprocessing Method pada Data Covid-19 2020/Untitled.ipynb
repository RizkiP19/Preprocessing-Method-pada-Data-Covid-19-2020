```python
# Data Reduction
features = ['Country/Region', 'Confirmed_x', 'Deaths_x', 'Recovered_x', 'Date']
data = data[features]
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d')
```

```python
monthly_data = data.resample('M', on='Date').sum()
```

```python
monthly_data['Cumulative_Cases'] = monthly_data['Confirmed_x'].cumsum()
monthly_data['Cumulative_Deaths'] = monthly_data['Deaths_x'].cumsum()
```

```python
monthly_data['Mortality_Rate'] = monthly_data['Deaths_x'] / monthly_data['Confirmed_x'] * 100
```
